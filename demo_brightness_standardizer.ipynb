```python
import staintools
%matplotlib inline
```

# i1

```python
i1 = staintools.read_image('./data/i1.png')
staintools.plot_image(i1, fig_size=6, title="Original Image")
```

```python
standardizer = staintools.BrightnessStandardizer()
i1_transformed = standardizer.transform(i1)
staintools.plot_image(i1_transformed, fig_size=6, title="Brightness Standardized Image")
```
